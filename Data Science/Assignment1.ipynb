```python
# Question 6

# b)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

age = [ '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
catagories = ['Smart Phone', 'Other Cell Phone', 'No Cell Phone']

counts = {
    'Smart Phone': [55, 70, 67, 32, 25, 18],
    'Other Cell Phone': [30, 26, 36, 32, 51, 24],
    'No Cell Phone': [2, 5, 7, 11, 17, 53]
}

data = {'age': age,
        'Smart Phone': counts ['Smart Phone'],
        'Other Cell Phone': counts ['Other Cell Phone'],
        'No Cell Phone': counts ['No Cell Phone']
        }

df = pd.DataFrame(data)

df_long = pd.melt(df, id_vars=['age'], var_name='catagories', value_name='count')

sns.barplot (x="catagories", y="count", hue="age", data=df_long)

plt.title('Cellphone Ownership for 25-34 Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
