```python
# Question 1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# a)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Fish.csv')
print(df.head())
print(repr(df.columns))
df.columns = df.columns.str.replace('\n', '').str.strip()
print(df.columns)
sns.scatterplot(x='Length (in cm)', y='Weight (in grams)', data=df)
plt.xlabel('Length (cm)')
plt.ylabel('Weight (g)')
plt.title('Scatterplot of Length vs Weight')
plt.show()

# Positive direction, linear form, strong correlation.

# b)
from scipy.stats import linregress
x = df["Length (in cm)"]
y = df["Weight (in grams)"]
model = linregress(x, y)
r = model.rvalue
print(f"Pearson correlation coefficient (r): {r:.4f}")

# There is a strong correlation, as it is close to 1.

# c)
df.loc[df["ID"] == 10, "Length"] = 49.2
model_new = linregress(df["Length (in cm)"], df["Weight (in grams)"])
r_modified = model_new.rvalue
print(f"Modified Pearson correlation coefficient (r): {r_modified:.4f}")

# d) No, the correlation did not change because the previous point was not an influential observation.
# This means that when we changed the point, it did not have much impact on the r.
```
