```python
# Question 6
import numpy as np
from scipy.stats import norm

# a)
proportion = norm.cdf(3)
print(f'Proportion below 3 standard deviations: {proportion:.5f}')

# b)
def stats_normal(n):
  sample = np.random.normal(loc=0, scale=1, size=n) # 1.
  sample_mean = np.mean(sample) # 2.
  sample_std = np.std(sample) # 3.
  threshold = sample_mean + 3 * sample_std #4.
  proportion = np.mean(sample < threshold)

  return {
      "mean": sample_mean,
      "std": sample_std,
      "proportion": proportion
  } #5.

# c)
np.random.seed(5)
def stats_normal(n):
  sample = np.random.normal(loc=0, scale=1, size=n)
  return np.mean(sample), np.std(sample)
result_50 = stats_normal(50)
result_500 = stats_normal(500)
result_5000 = stats_normal(5000)
result_50, result_500, result_5000

# d) As the sample size increases, the sample mean approaches 0 and the sample standard deviation approaches 1.
```
